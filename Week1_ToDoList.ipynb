```python
class Task:
    def __init__(self, description, due_date, priority):
        self.description = description
        self.due_date = due_date
        self.priority = priority
        self.completed = False

class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, description, due_date, priority):
        task = Task(description, due_date, priority)
        self.tasks.append(task)

    def display_tasks(self):
        if not self.tasks:
            print("No tasks in the To-Do List.")
        else:
            print("To-Do List:")
            for i, task in enumerate(self.tasks, start=1):
                status = "Completed" if task.completed else "Incomplete"
                print(f"{i}. {task.description} (Due: {task.due_date}, Priority: {task.priority}, Status: {status})")

    def mark_task_completed(self, task_index):
        if 0 <= task_index < len(self.tasks):
            self.tasks[task_index].completed = True

    def update_task(self, task_index, new_description, new_due_date, new_priority):
        if 0 <= task_index < len(self.tasks):
            task = self.tasks[task_index]
            task.description = new_description
            task.due_date = new_due_date
            task.priority = new_priority

    def remove_task(self, task_index):
        if 0 <= task_index < len(self.tasks):
            self.tasks.pop(task_index)

def add_task(to_do_list):
    description = input("Enter the description of the task: ")
    due_date = input("Enter due date on which the task has to be completed: ")
    priority = input("Enter priority of the task to be done: ")
    to_do_list.add_task(description, due_date, priority)

def display_tasks(to_do_list):
    to_do_list.display_tasks()

def mark_task_completed(to_do_list):
    task_index = int(input("Enter the task index to mark as completed: ")) - 1
    to_do_list.mark_task_completed(task_index)

def update_task(to_do_list):
    task_index = int(input("Enter the task index to update: ")) - 1
    n_description = input("Enter new description of the task: ")
    n_due_date = input("Enter new due date for completion of the task: ")
    new_priority = input("Enter new priority for the task: ")
    to_do_list.update_task(task_index, n_description, n_due_date, new_priority)

def remove_task(to_do_list):
    task_index = int(input("Enter the task index to be removed from the list: ")) - 1
    to_do_list.remove_task(task_index)

def quit_program(to_do_list):
    print("Successfully managed your To-Do List. Goodbye!")

def main():
    to_do_list = ToDoList()

    main_menu = {
        "1": add_task,
        "2": display_tasks,
        "3": mark_task_completed,
        "4": update_task,
        "5": remove_task,
        "6": quit_program,
    }

    while True:
        print("\nTo-Do List Menu:")
        print("1. Add Task")
        print("2. Display Tasks")
        print("3. Mark as Task Completed")
        print("4. Update Task")
        print("5. Remove Task")
        print("6. Quit")

        choice = input("Enter your choice: ")

        if choice in main_menu:
            main_menu[choice](to_do_list)
        else:
            print("Invalid input. Please try again.")

if __name__ == "__main__":
    main()
```
